```python
data['delta_renovated'] = (data['yr_renovated'] - data['yr_built']).astype(int).apply(lambda x: x if x > 0 else 0)
data.sample(10)
```

### Составьте несколько сводных таблиц

3.1 Найдите среднюю стоимость домов в зависимости от количества спален
Отсортируйте от меньшей стоимости к большей

3.2 Найдите минимальную, среднюю и максимальную стоимости домов в зависимости от состояния дома

3.3 Постройте таблицу с подсчетом количества домов в данных в зависимости от вида на набережную и оценкой вида

3.4 Каких домов в зависимости от этажности и количества спален больше?

```python
# Найдите среднюю стоимость домов в зависимости от количества спален Отсортируйте от меньшей стоимости к большей
data.groupby('bedrooms').agg({'price': 'mean'}).sort_values('price')
```

```python
# Найдите минимальную, среднюю и максимальную стоимости домов в зависимости от состояния дома
data.groupby('condition').agg({'price': ['min', 'mean', 'max']})
```

```python
#  Постройте таблицу с подсчетом количества домов в данных в зависимости от вида на набережную и оценкой вида
data.pivot_table(index='view',
               columns='waterfront',
               values='price',
               aggfunc='count',
              fill_value=0)
```

```python
#  Каких домов в зависимости от этажности и количества спален больше?
pd.crosstab(index=data['floors'],
            columns=data['bedrooms'])
```

Ответ: одноэтажных домов с 3 спальнями больше.

```python
# Постройте таблицу с подсчетом медианной стоимости домов в данных в зависимости от состояния дома и оценки дома
data.pivot_table(index='condition', # 1 способ
               columns='grade',
               values='price',
               aggfunc='median',
              fill_value=0)
```

```python
data.groupby(['grade', 'condition']).agg({'price': 'median'}).unstack('grade').fillna(0) # 2 способ
```
