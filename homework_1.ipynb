## Задача 2
На складе лежат разные фрукты в разном количестве.
Нужно написать функцию, которая на вход принимает любое количество названий фруктов и их количество, а возвращает общее количество фруктов на складе.

```python
def count_fruits(**kwargs):
    total = 0
    for value in kwargs.values():
        total += value
    return total

print("Общее количество фруктов:", count_fruits(banana=5, apple=5, lemon=10))
```

## Задача 3
Дан список с затратами на рекламу. Но в данных есть ошибки, некоторые затраты имеют отрицательную величину. 
Удалите такие значения из списка и посчитайте суммарные затраты
[100, 125, -90, 345, 655, -1, 0, 200].

Используйте list comprehensions

```python
data = [100, 125, -90, 345, 655, -1, 0, 200]
expenses_1 = [value for value in data if value >= 0] # 1 способ
expenses_2 = list(filter(lambda x: x >= 0, data)) # 2 способ
print(*expenses_1)
print(*expenses_2)
print("Затраты составляют:", sum(expenses_1))
print("Затраты составляют:", sum(expenses_2))
```

### Условие 4.2
Найдите выручку компании в зависимости от месяца.

Для этого напишите функцию, которая на вход принимает список с датами и список с выручкой, а на выходе словарь, где ключи - это месяцы, а значения - это выручка.
Используйте аннотирование типов.

```python
date =  ['2021-09-14', '2021-12-15', '2021-09-08', '2021-12-05', '2021-10-09', '2021-09-30', '2021-12-22', '2021-11-29',
         '2021-12-24', '2021-11-26', '2021-10-27', '2021-12-18', '2021-11-09', '2021-11-23', '2021-09-27', '2021-10-02', 
         '2021-12-27', '2021-09-20', '2021-12-13', '2021-11-01', '2021-11-09', '2021-12-06', '2021-12-08', '2021-10-09', 
         '2021-10-31', '2021-09-30', '2021-11-09', '2021-12-13', '2021-10-26', '2021-12-09'] 
revenue = [1270, 8413, 9028, 3703, 5739, 4095, 295, 4944, 5723, 3701, 4471, 651, 7037, 4274, 6275, 4988, 6930, 2971, 6592, 
        2004, 2822, 519, 3406, 2732, 5015, 2008, 316, 6333, 5700, 2887]

revenue_november = [revenue[i] for i in range(len(date)) if date[i][5:7] == '11'] # 4.1

print("Выручка компании в ноябре составляет", sum(revenue_november))
```

```python
def get_revenue_month(date: list[str], revenue: list[int]) -> dict: # 4.2
    month = [('01','январь'), ('02','февраль'), ('03','март'), ('04','апрель'), ('05','май'),('06','июнь'),
             ('07','июль'),('08','август'),('09','сентябрь'), ('10','октябрь'),('11','ноябрь'), ('12','декабрь')]
    for i in range(len(month)):
        for j in range(len(date)):
            if month[i][0] == date[j][5:7]:
                date[j] = month[i][1]
    revenue_dict = dict()
    for i, value in enumerate(revenue):
        revenue_dict[date[i]] = revenue_dict.get(date[i], 0) + value
    return revenue_dict
 
get_revenue_month(date, revenue)

    
```
